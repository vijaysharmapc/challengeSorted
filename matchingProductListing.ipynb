```python
# PANDAS WAY
import pandas as pd
```

```python
# had to do a pip install jsonline
# I initially thought of using just python, but then settled in favour of pandas
#this solution can be improved and made, more effecient
import json_lines
cnt =1
with open('products.txt', 'rb') as f:
    for item in json_lines.reader(f):
        if cnt == 1 :
            ds = pd.Series(item)
            df =pd.DataFrame(ds)
            prodDF = df.transpose()
            cnt += 1
        elif cnt == 2 :
            ds2 = pd.Series(item)
            df2 =pd.DataFrame(ds2)
            df2 = df2.transpose()
            prodDF = pd.concat([prodDF,df2])
```

```python
# had to do a pip install jsonline
# I initially thought of using just python, but then settled in favour of pandas
#this solution can be improved and made, more effecient
import json_lines
cnt =1
with open('listings.txt', 'rb') as f:
    for item in json_lines.reader(f):
        if cnt == 1 :
            ds = pd.Series(item)
            df =pd.DataFrame(ds)
            lstDF = df.transpose()
            cnt += 1
        elif cnt == 2 :
            ds2 = pd.Series(item)
            df2 =pd.DataFrame(ds2)
            df2 = df2.transpose()
            lstDF = pd.concat([lstDF,df2])
```

```python
new_lstDF= lstDF.replace({'_': ' '}, regex=True)
new_lstDF['manufacturer'] = new_lstDF['manufacturer'].str.lower()
new_lstDF['title'] = new_lstDF['title'].str.lower()
new_lstDF.head(5)
```

```python
new_prodDF= prodDF.replace({'_': ' '}, regex=True)
new_prodDF['product_name'] = new_prodDF['product_name'].str.lower()
new_prodDF['manufacturer'] = new_prodDF['manufacturer'].str.lower()
new_prodDF.head(5)
```

```python
new_lstDF[0:5]
```

```python
import sqlite3
conn = sqlite3.connect(':memory:')
new_lstDF.to_sql('new_lstDF', conn, index=False)
new_prodDF.to_sql('new_prodDF', conn, index=False)
query = """
SELECT [new_prodDF].[product_name],
       [new_prodDF].[manufacturer],
       [new_prodDF].[model],
       [new_lstDF].[title],
       [new_lstDF].[price],
       [new_lstDF].[manufacturer]
FROM   ([new_lstDF]
        INNER JOIN [new_prodDF]
                     ON trim([new_prodDF].[product_name]) LIKE trim([new_lstDF].[title])
)
"""
df3 = pd.read_sql_query(query, conn)
conn.close()
#        LEFT OUTER JOIN [new_prodDF]
```
